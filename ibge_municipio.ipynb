```python
from urllib.request import Request, urlopen
import csv
from bs4 import BeautifulSoup
from pathlib import Path
from collections import defaultdict
import unidecode
```

```python
rows = []

with open(str(Path.cwd()) + '/cidades.csv', newline='') as f:        
    reader = csv.DictReader(f, delimiter=';')
    for row in reader: 
         rows.append(row)    
```

```python
#request das páginas
url_ibge = 'https://www.ibge.gov.br/cidades-e-estados'
paginas = []

for row in rows: 
    print(f"{row['nome_cidade']} - {row['uf']}")
    url = f"{url_ibge}/{row['uf']}/{unidecode.unidecode(row['nome_cidade'])}.html".replace(' ','-').replace("'", "") 
    print(url)
    link = Request(url, headers={'User-Agent':'Mozilla/5.0'})
    paginas.append([row['nome_cidade'], row['uf'],urlopen(link).read().decode('utf-8', 'ignore')])
```

```python
#raspagagem dos dados nas páginas

for pagina in paginas:
    print(pagina[0] + '-' + pagina[1] + '#'*100)
    soup = BeautifulSoup(pagina[2], "html.parser")
    divs = soup.find_all("div", {"class": "indicador"})
    for div in divs:
        div_label =div.find("div","ind-label")
        label = div_label.find("p")
        print(label.text + '=')
        div_value = div.find("div", "ind-value")
        values = div.find("p","ind-value")
        for value in values.children:
            print(value.text)
```
